```python
import os
import string


def analyze_password(password: str) -> dict:
    """
    Evaluates the strength of a password based on length and complexity.
    """
    score = 0
    suggestions = []

    # 1. Check length
    if len(password) >= 8:
        score += 1
    else:
        suggestions.append("Make the password at least 8 characters long.")

    # 2. Check lowercase
    if any(c.islower() for c in password):
        score += 1
    else:
        suggestions.append("Add lowercase letters.")

    # 3. Check uppercase
    if any(c.isupper() for c in password):
        score += 1
    else:
        suggestions.append("Add uppercase letters.")

    # 4. Check digits
    if any(c.isdigit() for c in password):
        score += 1
    else:
        suggestions.append("Add numbers.")

    # 5. Check special symbols
    special_characters = string.punctuation
    if any(c in special_characters for c in password):
        score += 1
    else:
        suggestions.append("Add special characters (e.g., !@#$%).")

    # Determine strength label
    if score == 5:
        strength = "Strong"
    elif score >= 3:
        strength = "Medium"
    else:
        strength = "Weak"

    return {
        "strength": strength,
        "score": score,
        "length": len(password),
        "suggestions": suggestions
    }


def main():
    used_passwords_file = "used_passwords.txt"

    # Load previously used passwords if the file exists
    used_passwords = set()
    if os.path.exists(used_passwords_file):
        with open(used_passwords_file, "r") as f:
            used_passwords = set(line.strip() for line in f if line.strip())


    print("=== Password Strength Analyzer ===")

    while True:
        password = input("\nEnter a password to analyze (or 'q' to quit): ")
        if password.lower() == 'q':
            print("Exiting...")
            break

        if not password:
            print("Password cannot be empty. Please try again.")
            continue

        # Check against reused passwords
        if password in used_passwords:
            print("\n[!] Warning: This password was used before; avoid reusing old passwords.")

        # Analyze the new password
        result = analyze_password(password)

        # Display results
        print("\n--- Analysis Result ---")
        print(f"Strength:  {result['strength']}")
        print(f"Score:     {result['score']}/5")
        print(f"Length:    {result['length']}")

        if result['suggestions']:
            print("Suggestions for improvement:")
            for sug in result['suggestions']:
                print(f"  - {sug}")

        # Prompt to save as a used password (preventing future reuse)
        if password not in used_passwords:
            save_ans = input("\nDo you want to save this password so it is not reused later? (y/n): ").strip().lower()
            if save_ans == 'y':
                used_passwords.add(password)
                with open(used_passwords_file, "a") as f:
                    f.write(password + "\n")
                print("Password saved to used passwords lists.")


if __name__ == "__main__":
    main()
```
